```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_name = "deepseek-ai/deepseek-coder-6.7b-base"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.bfloat16, device_map="auto")
model.eval()
```

```python
input_text = "details comparision betweeen chat gpt o1 vs deep seek r3 include some good examples digrams in 10k words"
```

```python
inputs = tokenizer(input_text, return_tensors="pt").to("cuda")
```

```python
outputs = model.generate(**inputs, max_new_tokens=500, temperature=0.7, top_p=0.95, do_sample=True)
```

```python
result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(result)
```
